```r
library(forecast)
```

```r
data <- read.csv("./data.csv", header = TRUE)
plot(data ,type = "l", col = "blue",lwd = 2,main = "Gasoline")
```

Параметры оценены из графиков частичной автокорреляции (файл с python и предыдущие задания).

```r
p <- 2
P <- 0
nnet <- nnetar(data$gasoline[1:248], p=p, P=P, size=10)
```

```r
fcast <- forecast(nnet, h=12)
plot(fcast)
```

```r
mean((fcast$mean[1:12] - data$gasoline[249:260])**2)
```
